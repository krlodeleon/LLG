```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os
import time

current_dir = os.getcwd()
package_dir = current_dir + '/packages'
sys.path.append(package_dir)

import rydberg as rd
```

# Generating system

```python
N = 100
n_networks = 10
n_thermalSamples = 100
n_samples = n_networks*n_thermalSamples
K = 1

systems = rd.MagneticSystem(N, n_samples, K=K) # generating system
systems.align_m() # randomizing initial spin config. m0
systems.m = np.broadcast_to(systems.m[0], (n_samples, N, 3)) # copying one m0 throughout all samples

# setting Js of thermal samples batches with one J
J_toSet = []
for J in systems.J[:n_networks]:
    J_toSet.append(np.broadcast_to(J, (n_thermalSamples, N, N)))
J_toSet = np.array(J_toSet)
J_toSet = J_toSet.reshape(n_samples, N, N)
systems.J = J_toSet
del J_toSet # deleting buffer
```

# Running dynamics

```python
dt = 1
n_burnSteps = 1000
n_steps = 1000
T = 0.1

steps = np.arange(n_steps)

# storing initial spins
m0 = systems.m
m0 = m0.reshape(n_networks, n_thermalSamples, N, 3)
m0 = np.broadcast_to(m0[np.newaxis, :, :, :, :], (n_steps, n_networks, n_thermalSamples, N, 3))

# burning-in system
for i in range(n_burnSteps):
    systems.LLG_1stepHeun(dt, alpha=1, precession=0, T=T)

history = []
for i in range(n_steps):
    history.append(systems.m)
    systems.LLG_1stepHeun(dt, alpha=1, precession=0, T=T)
    print(i)
history = np.array(history).reshape(n_steps, n_networks, n_thermalSamples, N, 3)
print('Done.')
```

# Correlation

```python
# calculating correlation
corr = np.mean(history*m0, axis=(1,2,3))
```

```python
l,r = 1, None
plt.plot(steps[l:r], corr[l:r,2])
```

```python
l,r = 1, None
plt.plot(np.log(steps[l:r]), np.log(corr[l:r,2]**2))
```
